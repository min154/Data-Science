## 4-2. 서울시 구별 주유소 가격 정보 얻기

```python
#사이트가 리뉴얼된 것으로 추정! 교재의 주소 대신 '싼 주유소 찾기' 주소로 연결해서 실행
driver = webdriver.Chrome()
driver.get("https://www.opinet.co.kr/searRgSelect_1508.do")
```

```python
from selenium.webdriver.common.by import By
gu_list_raw = driver.find_element("xpath", """//*[@id="SIGUNGU_NM0"]""")
gu_list = gu_list_raw.find_elements(By.TAG_NAME, "option")
```

```python
gu_names = [option.get_attribute("value") for option in gu_list]
gu_names.remove('')
gu_names
```

```python
element = driver.find_element("id", "SIGUNGU_NM0")
element.send_keys(gu_names[0])
```

```python
xpath = """//*[@id="searRgSelect"]"""
element_sel_gu = driver.find_element("xpath", xpath).click()
```

```python
xpath = """//*[@id="glopopd_excel"]"""
element_get_excel = driver.find_element("xpath", xpath).click()
```

```python
#코드 실행하고 창 띄워두면 자동으로 시군구가 바뀌며 지도를 보여주는 걸 볼 수 있음

import time
from tqdm import tqdm_notebook

for gu in tqdm_notebook(gu_names):
    element = driver.find_element("id", "SIGUNGU_NM0")
    element.send_keys(gu)
    
    time.sleep(2)
    
    xpath = """//*[@id="searRgSelect"]"""
    element_sel_gu = driver.find_element("xpath", xpath).click()
    
    time.sleep(1)
    
    xpath = """//*[@id="glopopd_excel"]"""
    element_get_excel = driver.find_element("xpath", xpath).click()
    
    time.sleep(1)
```

```python
driver.close()
```

## 4-3. 구별 주유 가격에 대한 데이터의 정리

```python
import pandas as pd
from glob import glob
```

```python
glob('../data/지역*.xls')
```

```python
stations_files = glob('../data/지역*.xls')
```

```python
tmp_raw = []

for file_name in stations_files:
    tmp = pd.read_excel(file_name, header = 2)
    tmp_raw.append(tmp)
    
station_raw = pd.concat(tmp_raw)
```

```python
station_raw.info()
```

```python
station_raw.head()
```

```python
stations = pd.DataFrame({'Oil_store' : station_raw['상호'],
                        '주소' : station_raw['주소'],
                        '가격' : station_raw['휘발유'],
                        '셀프' : station_raw['셀프여부'],
                        '상표' : station_raw['상표']})
stations.head()
```

```python
stations['구'] = [eachAddress.split()[1] for eachAddress in stations['주소']]
stations.head()
```

```python
stations['구'].unique()
```

```python
stations[stations['가격'] == '-']
```

```python
stations = stations[stations['가격'] != '-']
stations.head()
```

```python
stations['가격'] = [float(value) for value in stations['가격']]
```

```python
stations.reset_index(inplace = True)
del stations['index']
```

```python
stations.info()
```

## 4-4. 셀프 주유소는 정말 저렴한지 boxplot으로 확인하기

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import platform

path = "c:/Windows/Fonts/malgun.ttf"
from matplotlib import font_manager, rc
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~') 
```

## 4-5. 서울시 구별 주유 가격 확인하기

```python
import json
import folium
import googlemaps
import warnings
warnings.simplefilter(action = "ignore", category = FutureWarning)
```

```python
stations.sort_values(by = '가격', ascending = False).head(10)
```

```python
stations.sort_values(by = '가격', ascending = True).head(10)
```

```python
import numpy as np
gu_data = pd.pivot_table(stations, index = ["구"], values = ["가격"], aggfunc = np.mean)
gu_data.head()
```

```python
geo_path = '../data/05. skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))

map = folium.Map(location=[37.5502, 126.982], zoom_start=10.5, 
                 tiles='Stamen Toner')

map.choropleth(geo_data = geo_str,
               data = gu_data,
               columns=[gu_data.index, '가격'],
               key_on='feature.id')

map
```
